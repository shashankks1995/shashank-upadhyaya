## k-means is used here , and by using k-means elbow techinique we can find no of cluster

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
k_range = range(1, 10)
inertias = [] 
for k in k_range:
    model = KMeans(n_clusters=k)# 
    model.fit(PCA_df.iloc[:,:3])
    inertias.append(model.inertia_)     
plt.plot(k_range, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(k_range) 
plt.show()
```

## Randomforest classifier and xg boostclassifier is used

```python
from xgboost import XGBClassifier
```

```python
model_xg=XGBClassifier(learning_rate=0.15,random_state=200,n_estimators=40,gamma=0.25)
model_xg.fit(X_train,y_train)
y_pre_xg=model_xg.predict(X_test)
```

```python
pd.crosstab(y_test,y_pre_xg)
```
